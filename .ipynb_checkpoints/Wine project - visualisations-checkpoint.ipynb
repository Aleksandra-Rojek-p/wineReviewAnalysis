```python
grapes = df[["winery", "variety"]].groupby(by = "winery").count()
max_num = max(grapes["variety"])
min_num = min(grapes["variety"])

k = 0

print("Max number of grapes: ",max_num)

#number of wineries using max number of types of grapes
for i in range(len(grapes)):
    if(grapes["variety"][i] == max_num ):
        k = k + 1
print(str(k), "winery(ies) use this number of grapes")



k = 0

print("Min number of grapes: ", min_num)

#number of wineries using min number of types of grapes
for i in range(len(grapes)):
    if(grapes["variety"][i] == min_num ):
        k = k + 1

print(str(k), "winery(ies) use this number of grapes")
```

```python
data_to_plot = df.groupby(['country',]).mean()['price'].sort_values(ascending=False).to_frame()

plt.figure(figsize=(15,13))
sns.pointplot(x = data_to_plot.price, y = data_to_plot.index, color='maroon', orient='h', markers='o')
plt.title('Average price of wine by country')
plt.xlabel('Price ($)');
```

```python
data_to_plot = df.groupby(['country',]).mean()['points'].sort_values(ascending=False).to_frame()

plt.figure(figsize=(15,13))
sns.pointplot(x = data_to_plot.points ,y = data_to_plot.index, color='maroon', orient='h')
plt.title('Average wine rating by country')
plt.xlabel('Points');
```
